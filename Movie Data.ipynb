```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts configuration of created plots

# Read in the data

df = pd.read_csv(r'C:\Users\jstott\Desktop\Data Analysis Stuff\CSV Files\movies.csv\movies.csv')
```

```python
# Look at the data

df.head()
```

```python
# Look for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for our columns

df.dtypes
```

```python
# Dropping missing values
df = df.dropna()
```

```python
# Changing the data type of the columns
df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
# Creating correct year column based off released column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
# Ordering by gross revenue

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',100)
```

```python
# Drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Prediction: High Budget =  High Gross
# Prediction: Company correlates to gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget VS Gross Earnings')
plt.xlabel('Gross Earnings (In Millions)')
plt.ylabel('Budget for Film (In Millions)')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Correlation between columns

df.corr()
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()

plt.show
```

```python
# Looks at Company

df_numeral = df

for col_name in df_numeral.columns:
    if(df_numeral[col_name].dtype == 'object'):
        df_numeral[col_name] = df_numeral[col_name].astype('category')
        df_numeral[col_name] = df_numeral[col_name].cat.codes
    
df_numeral
```
